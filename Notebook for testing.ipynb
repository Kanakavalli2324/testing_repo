```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
from delta.tables import DeltaTable
from pyspark.sql.window import Window
```

```python
spark=SparkSession.builder.appName('SparkByExamples.com').config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension").config("spark.sql.catalog.delta", "io.delta.sql.catalog.DeltaCatalog").getOrCreate()
```

```python
columns=['id','name','age']
data=[(1,'John',20),(2,'Mary',30),(3,'Bob',40)]
df=spark.createDataFrame(data=data,schema=columns)
df.write.format('delta').mode('overwrite').saveAsTable("dummyTable")
```
